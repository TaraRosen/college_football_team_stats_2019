```python
# splitting the team column into Team and Conference columns

# new data frame with split value columns 
new = df['Team'].str.split('(', n = 1, expand = True) 
  
# making separate team column from new data frame 
df['Team']= new[0] 
  
# making Conference column from new data frame 
df['Conference']= new[1] 
    
# df display 
df.head()
```

```python
# remove ')' from Conference values

df['Conference'] = df['Conference'].map(lambda x: x.rstrip(')'))
```

```python
df.head()
```

```python
# splitting the win-loss column into win and loss columns

# new data frame with split value columns 
new = df['Win-Loss'].str.split('-', n = 1, expand = True) 
  
# making separate win column from new data frame 
df['Win']= new[0] 
  
# making loss column from new data frame 
df['Loss']= new[1] 
  
# Dropping old Name columns 
df.drop(columns =['Win-Loss'], inplace = True) 
  
# df display 
df.head()
```

```python
# move Loss column to front

df = df[['Loss'] + [col for col in df.columns if col != 'Loss']]
```

```python
# move Win column to front

df = df[['Win'] + [col for col in df.columns if col != 'Win']]
```

```python
# move Team column to front

df = df[['Team'] + [col for col in df.columns if col != 'Team']]
```

```python
# move Division column to front

df = df[['Conference'] + [col for col in df.columns if col != 'Conference']]
```
